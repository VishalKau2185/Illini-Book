```python
def gradient_descent(m1_now, m2_now, m3_now, b_now, points, L):
  m1_gradient = 0
  m2_gradient = 0
  m3_gradient = 0
  b_gradient = 0

  n = len(points)

  for i in range(n):
    x1 = points.iloc[i].TGTS
    x2= points.iloc[i].RECYDS
    x3= points.iloc[i].CATCHPER
    y = points.iloc[i].FPPG

    m1_gradient += -(2/n) * x1 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    m2_gradient += -(2/n) * x2 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    m3_gradient += -(2/n) * x3 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    b_gradient += -(2/n) * (y - (m1_now * x1 + m2_now * x2 + b_now))

  m1 = m1_now - m1_gradient * L
  m2 = m2_now - m2_gradient * L
  m3 = m3_now - m3_gradient * L
  b = b_now - b_gradient * L
  return m1, m2, m3, b
```

```python
m1 = 0
m2 = 0
m3 = 0
b = 0
L = 0.0001
epochs = 100

for i in range(epochs):
  m1, m2, m3, b = gradient_descent(m1, m2, m3, b, players, L)
print(m1, m2, m3, b)
```

```python
def calculate_result(m1, m2, m3, b, points):
  n = len(points)

  array = []
  for i in range(n):
    x = points.iloc[i].TGTS
    x2 = points.iloc[i].RECYDS
    x3 = points.iloc[i].CATCHPER
    array.append(m1 * x + m2 * x2 + m3 * x3 + b)
  return array
```

```python
players_2 = pd.read_csv(io.BytesIO(uploaded['wr_stats_2022.csv']))
players_2.rename(columns = {'REC YDS':'RECYDS'}, inplace = True)
players_2.rename(columns = {'CATCH %':'CATCHPER'}, inplace = True)
```

```python
result = calculate_result(m1, m2, m3, b, players_2)
```

```python
players_2['Predictions'] = result
players_2.head(50)
```

```python
def percentage_change(col1,col2):
    return ((col2 - col1) / col1) * 100

players_2['error'] = -1 * percentage_change(players_2['FPPG'],players_2['Predictions'])
average_error = players_2['error'].mean()
average_error
```
